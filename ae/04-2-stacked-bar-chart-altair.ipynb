# Stacked bar chart

## Setup


```python
import pandas as pd
import altair as alt
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)
alt.data_transformers.disable_max_rows()
```

## Data

```python
ROOT = "https://raw.githubusercontent.com/kirenz/datasets/master/"
DATA = "loans.csv"

df = pd.read_csv(ROOT + DATA)

df.homeownership = df.homeownership.astype("category")
df.application_type = df.application_type.astype("category")
```

## Chart

```python
alt.Chart(df).mark_bar().encode(
    x=alt.X('homeownership').sort('-y').axis(
        title="Homeownership", 
        titleAnchor="start", 
        labelAngle=0),
    y=alt.Y('homeownership', aggregate='count').axis(
        title = "Count", 
        titleAnchor="end",
        grid=False),
    color= alt.Color('application_type').legend(
        title="Type",
        orient="right")
).properties(
    title='This is a stacked bar chart',
    width=400,
    height=300
).configure_title(
    fontSize=20,
    font='Arial',
    anchor='start',
    color='black'
).configure_axis(
    labelFontSize=14,
    titleFontSize=16
).configure_view(strokeWidth=0)
```
