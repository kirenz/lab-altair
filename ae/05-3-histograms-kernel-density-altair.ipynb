# Histograms

## Setup

```python
import pandas as pd
import altair as alt
import numpy as np
import random
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)
alt.data_transformers.disable_max_rows()
```

## Data

```python
# Create data with normal distribution

random.seed(123)

mean, sd = 11, 2
interest_rates_normal = np.random.normal(mean, sd, 1000)

df_normal = pd.DataFrame({"interest_rate": interest_rates_normal})
```

## Analysis

### Histogram

```python
# Histogram 
alt.Chart(df_normal).mark_bar().encode(
    x=alt.X("interest_rate").bin(),
    y=alt.Y("interest_rate", aggregate="count")
)
```

### Kernel density estimate plot

```python
 # Kernel density estimate (KDE) plot
alt.Chart(df_normal).transform_density(
    'interest_rate',
    as_=['interest_rate', 'density'],
).mark_area().encode(
    x="interest_rate",
    y='density:Q',
)
```
